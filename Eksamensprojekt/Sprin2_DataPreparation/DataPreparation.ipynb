#### Visualization

```python
import numpy as np

data_column_category = df.select_dtypes(exclude=[np.number]).columns

data_column_category

from sklearn.preprocessing import LabelEncoder

#Creating the object instance

label_encoder = LabelEncoder()

for i in data_column_category:

    df[i] = label_encoder.fit_transform(df[i])

print("Label Encoded Data: ")

df.head()
```

```python
# Histogram
```

```python
df.hist(figsize=(22,22))
```

```python
# Group by SMOKING and LUNG_CANCER and count occurrences
count_data = df.groupby(['SMOKING', 'LUNG_CANCER']).size().reset_index(name='Count')

# Plot
sns.set_style("whitegrid")

sns.barplot(x='SMOKING', y='Count', hue='LUNG_CANCER', data=count_data)
plt.title('Count of Lung Cancer by Smoking Status')
plt.xlabel('Smoking')
plt.ylabel('Count')
plt.legend(title='Lung Cancer')
plt.xticks(rotation=0)
plt.show()

# 1 = Non smokers
# 2 = Smokers
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.boxplot(x='SMOKING', y='LUNG_CANCER', data=count_data)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
combined_data = df.groupby('SMOKING')['LUNG_CANCER'].mean().plot(kind='bar')
plt.title('Average Lung Cancer by Smokers and Non Smokers')
plt.ylabel('Average lung cancer by type')

avg_quality = df.groupby('SMOKING')['LUNG_CANCER'].mean()
quality_difference = abs(avg_quality[1] - avg_quality[2])
#print(f'The Average Quality differnce between Red and White Wine: {quality_difference}')
plt.show()
```
